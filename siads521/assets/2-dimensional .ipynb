# Two-dimensional 

In the previous lecture, we talked about the matplolit architecture. At this point, you should know the fundamental of matplotlit. In the beggining of the course, I am going to introduce you a serveral of two-dimensional visisulizatiosns with a new python libaray "seaborn.

```python
## What is seaborn? Seaborn is a Python data visualization library based on matplotlib.
##It provides a high-level interface for drawing attractive and informative statistical graphics.


## Let's import seaborn
import seaborn as sns 
import matplotlib.pyplot as plt
import pandas as pd
```

https://seaborn.pydata.org/

```python
## let's load the dataset "iris"
df = sns.load_dataset('iris')
## Copy df dataframe into a new variable called df1, so we can pop up the column "species" for clustermap.
df1=df.copy()
species = df1.pop("species")
df.head()
```

```python
## The regression line of setosa is higher than versicolor and virginica. 

sns.lmplot(x='sepal_length',y='sepal_width',hue="species",data=df)
```

```python
# What if we want to see the distribution of the graph? 

sns.jointplot(x='sepal_length',y='sepal_width',data=df)
```

```python
## It's a bit difficult to see where the interesting areas in the plot are,
## so it's worth trying a hexbin plot. Go ahead and copy the above code block 
## and add kind="hex" to the jointplot parameters.
sns.jointplot(x='sepal_length',y='sepal_width',data=df, kind ='hex')
```

```python
## What if we want to see the density of variables x and y? We can change kind to "Kde".

sns.jointplot(x='sepal_length',y='sepal_width',data=df,kind="kde")
```

```python
## What if we want to show all numeric values regarding setosa? Let's do the pairplot and
## query setosa out of a variable called species. 
sns.pairplot(df.query("species == 'setosa'"))
```

```python
## What if we want to see all the numeric values regarding each different species? We can use
## hue to do so. 
sns.pairplot(df,hue="species")
```

```python
## If we want to demonstrate the data in a hierarchically-clustered heatmap, clustermap is 
## a good option. The y-axis is the index, and the legend are numeric values of
## lengths or widths. 
a = sns.clustermap(df1)
```

```python
## First of all, let's load the fmri data

fmri = sns.load_dataset("fmri")
```

```python
fmri.head()
```

```python
## Let's take a look at some basic stats on timepoint and signal columns 

fmri.describe()
```

```python
## It shows the regression line with the standard deviation area. 

sns.relplot(x="timepoint", y="signal", kind="line",  ci="sd", data=fmri);
```

```python
## What if we want to just show the regression line?
sns.relplot(x="timepoint", y="signal", kind="line", ci=None, data=fmri);
```

```python
## If we do want to plot the resulting estimate, set estimator to None. 
sns.relplot(x="timepoint", y="signal", kind="line", estimator=None, data=fmri);
```

```python
## What if we want to see the regressions line from different events? Set hue to 'event'
sns.relplot(x = "timepoint", y = "signal", kind = "line", data = fmri, hue = "event");
```

```python
## If we want to see how event had been doing in different region, then we can add sytle in the paramter.
sns.relplot(x="timepoint", y="signal", kind="line", data=fmri, hue="region", style="event");
```

```python
## Let's load the stock dataset
AABA = pd.read_csv('assets/AABA.csv')
```

```python
AABA.head(5)
AABA.info()
```

```python
AABA['Day']=[d.split('-')[2] for d in AABA.Date]
AABA.head(5)
```

```python
AABA1=AABA.copy()
Name = AABA1.pop("Name")
AABA1.head()
```

```python
## select the first 10 rows in the dataset
AABA12=AABA1[:20]
AABA12
```

```python
# Initialize the figure
plt.style.use('seaborn-darkgrid')
 
# create a color palette
palette = plt.get_cmap('Set1')

fig= plt.figure(figsize=(50,20))

# multiple line plot
num=0
for column in AABA1.drop(['Date','Day','Volume'], axis=1):
    num+=1
    
    # Find the right spot on the plot
    plt.subplot(2,2, num)
 
    # Plot the lineplot
    plt.plot(AABA1['Day'], AABA1[column], marker='', color=palette(num), linewidth=1.9, alpha=0.9, label=column)
 
    # Not ticks everywhere
    if num in range(20) :
        plt.tick_params(labelbottom='off')

 
    # Add title
    plt.title(column, loc='left', fontsize=12, fontweight=0, color=palette(num) )
 
# general title
plt.suptitle("", fontsize=13, fontweight=0, color='black', style='italic', y=1.02)
 
# Axis title
plt.text(1, 1.82, '', ha='center', va='center')
plt.text(1, 1.82, '', ha='center', va='center', rotation='vertical')


## As we can see, these are bad visualizations since there are too many data points and it is hard to tell the line 
```
